```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vector field (dx/dt, dy/dt)
def vector_field(X, Y):
    dx = X - Y
    dy = X + Y
    return dx, dy

# Create a grid of points
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)

X, Y = np.meshgrid(x, y)

# Calculate the vector field at each point on the grid
DX, DY = vector_field(X, Y)

# Normalize the vectors for better visualization
magnitude = np.sqrt(DX**2 + DY**2)
DX_normalized = DX / magnitude
DY_normalized = DY / magnitude

# Create the phase portrait
plt.figure(figsize=(7, 7))
plt.quiver(X, Y, DX_normalized, DY_normalized, magnitude, pivot='mid', cmap=plt.cm.viridis)
plt.title('Phase Portrait')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
