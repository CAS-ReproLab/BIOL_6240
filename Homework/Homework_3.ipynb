# The pH scale.

The pH scale is a logarithmic scale used to specify the degree of auto-ionization of water. Water is important in biochemistry because of its micro and macroscopic properties including amphotericity, polarity, density anomolies, and colligative properties which ultimately influence its behavior as a `universal solvent`. Molecules either dissolve well in water, or they don't, and both of those types of molecules serve important functions in cells. When the solvent properties of water change, it can have dramatic effects on the stability and behavior of biomolecules. 

- **Acidic solutions:** Have higher $[H^+]$ and thus lower pH (pH < 7).
- **Basic (alkaline) solutions:** Have lower $[H^+]$ and thus higher pH (pH > 7).


## Definition of pH

`pH is a bulk property of water`. The pH of a solution is defined as the negative logarithm (base 10) of the hydrogen ion concentration:

$$ pH = -\log_{10} [H^+] $$

where $[H^+]$ is the concentration of hydrogen ions in moles per liter (M).


## Derivation

The derivation of the pH scale is based on the ionic dissociation of pure water into hydronium and hydroxide:

$$ H_2O \leftrightarrow H^+ + OH^- $$

At 25Â°C (298 K), the auto-ionization (equilibrium) constant ($K_w$) for this reaction is:

$$ K_w = [H^+][OH^-] = 1.0 \times 10^{-14} \, \text{M}^2 $$

For neutral water at this temperature:

$$ [H^+] = [OH^-] = \sqrt{K_w} = 1.0 \times 10^{-7} \, \text{M} $$

Thus, the pH of neutral water is the negative logarithm base 10 of the hydronium concentration:

$$ pH = -\log_{10} [1.0 \times 10^{-7}] = 7.0 $$

## Importance in Biochemistry

- **Enzyme Activity:** Most enzymes have an optimal pH range. Deviations from this range can alter the enzyme's structure and function.
- **Protein Structure:** pH affects the ionization of amino acids, influencing protein structure and interactions through solubility, active site chemistry (for enzymes), and protein folding.
- **Metabolic Processes:** Many metabolic processes are pH-dependent and are sensitive to changes in intracellular pH.
- **Reactivity of resonance stabilized groups:** Recall from our discussion in class that protonated V. deprotonated carboxylic acid groups have different resonance statibility.

## Cellular Scale Volumes and the pH Scale

In class we discussed the limitations of the pH scale as the volume of water gets increasingly smaller, as in the case of subcellular organelles which can have luminal volumes near the attoliter scale ($10^{-18} L$). 

Below is a program that calculates and plots the minimal pure water volume at which the number of protons (calculated for $25\degree Celsius$ becomes discrete, meaning $\leq 1$.

````python
import numpy as np
import matplotlib.pyplot as plt


avogadro_number = 6.022e23  
h_concentration = 1e-3  
volume_range = np.logspace(-24, -1, 100)  


h_ions = h_concentration * avogadro_number * volume_range


pH = -np.log10(h_concentration * volume_range)

plt.figure(figsize=(8, 6))
plt.plot(volume_range, pH, label='pH', color='b')


discrete_points = h_ions < 1
plt.scatter(volume_range[discrete_points], pH[discrete_points], color='r', label='Discrete behavior (H+ < 1 ion)')

plt.xscale('log')
plt.xlabel('Volume (L)')
plt.ylabel('pH')
plt.title('pH as a Function of Volume - Breakdown at Small Volumes')
plt.axvline(np.max(volume_range[discrete_points]), color='r', linestyle='--', label='Single molecule limit')
plt.legend()

plt.show()
````

# pKa and The Protonation State of Amino Acid Functional Groups:

Strong acids/bases will dissociate completely and the pH can be calculated from the $k_{w}$ and the total $[H^+]$ and $[OH^-]$. Weak acid/base dissociation is not complete, and occurs as a `proportion` of the dissociated/undissociated forms. For weak acid/base solutions, the pH can be estimated using the Henderson-Hasselbalch equation:

$$ pH = pKa + \log_{10} \left( \frac{[A^-]}{[HA]} \right) $$

where:
- pH is the acidity of the solution.
- $pK_a$ is the acid dissociation constant, a specific property of each amino acid (there may be more than one if there are multiple ionizable groups).
- $[{A}^-]$ is the concentration of the deprotonated form of the amino acid.
- $[{HA}]$ is the concentration of the protonated form of the amino acid.

The $pKa$ is a constant under specified conditions, and represents the dissociation state at which $[H^+] = [OH^-]$. In other words, the pKa is the pH at which there is a 1:1 concentration of dissociated/undissociated forms.

This equation is particularly useful for designing buffer systems that maintain a constant pH in biological experiments. Buffers are effectively a thermodynamic `sink` for protons. 


The program below estimates the proportion of protonated amino acid molecules $[{HA}]$ to deptoronated $[A^-]$ in a solution at a given pH and pKa. The program plots the Henderson-Hasselbalch equation with pH on the y-axis and the incremental addition of concentrated based $[{A}^-]$ (e.g., NaOH which raises pH) on the x-axis.

Note:

- I needed to rearrange the Henderson-Hasselbalch equation to solve for $[HA]$.
- I Used numpy and matplotlib for calculations and plotting.


````python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

def estimate_protonation(pKa):
    """
    Estimate the proportion of protonated amino acid molecules in a solution at a given pH and pKa.
    :param pKa: Acid dissociation constant of the amino acid.
    :param pH_range: A range of pH values.
    :return: Plot showing the relationship between pH and the proportion of protonated amino acid.
    """
    # Calculate the proportion of protonated amino acid (HA) using the rearranged Henderson-Hasselbalch equation:
    # [HA] = 1 / (1 + 10^(pH - pKa))
    pH_range = np.linspace(0, 14, 100)  # creates an uniformly distributed array of 100 pH values from 0 to 14
    proportion_HA = 1 / (1 + 10 ** (pH_range - pKa))

    return pH_range, proportion_HA 


pKa = 9  # Example pKa for an amino acid

pH_range, proportion_HA = estimate_protonation(pKa)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(pH_range, proportion_HA, label=f'pKa = {pKa}')
plt.axvline(x=pKa, color='r', linestyle='--', label='pKa')
plt.title('Protonation State of Amino Acid vs. pH')
plt.xlabel('pH')
plt.ylabel('Proportion of Protonated Amino Acid (HA)')
plt.yticks(np.arange(0, 1.1, 0.1)) # allows y-axis to have increments of 0.1
plt.legend()
plt.grid(True)
plt.show()
````

# Extending the Concept of pKa to Alanine: An Amino Acid with Two Ionizable Groups

Alanine is one of the simplest amino acids with the following structure:

- An **Î±-amino group** ($-\text{NH}_3^+$)
- An **Î±-carboxyl group** ($-\text{COOH}$)
- A **methyl side chain** (non-ionizable)

Each ionizable group has its own pKa value because they can donate or accept protons independently. For alanine:

- **pKaâ (Î±-carboxyl group)**: Approximately **2.34**
- **pKaâ (Î±-amino group)**: Approximately **9.69**

Here, we'll extend the concept of pKa to alanine by:

1. Applying the Henderson-Hasselbalch equation to each ionizable group.
2. Calculating the fraction of protonated and deprotonated forms.
3. Determining the net charge at different pH levels.
4. Visualizing these relationships with Python programming.

First, we'll apply the Henderson-Hasselbalch Equation, which as we discussed above, relates pH, pKa, and the ratio of the concentrations of the deprotonated and protonated forms of an ionizable group:

$$
\text{pH} = \text{pKa} + \log\left( \frac{[\text{A}^-]}{[\text{HA}]} \right)
$$

For alanine:

- **Î±-Carboxyl Group**:

  $$
  \text{COOH} \leftrightarrows \text{COO}^- + \text{H}^+
  $$

  $$
  \text{pH} = \text{pKa}_1 + \log\left( \frac{[\text{COO}^-]}{[\text{COOH}]} \right)
  $$

- **Î±-Amino Group**:

  $$
  \text{NH}_3^+ \leftrightarrows \text{NH}_2 + \text{H}^+
  $$

  Since the amino group accepts a proton, the equation is adjusted:

  $$
  \text{pH} = \text{pKa}_2 + \log\left( \frac{[\text{NH}_2]}{[\text{NH}_3^+]} \right)
  $$

Next, we'll calculate the fractional concentrations. Remember, the pKa is the pH at which $[H]:[HA]$ is $1:1$...

**Î±-Carboxylic acid Group**:

- **Fraction of Protonated Form** ($f_{\text{COOH}}$):

  $$
  f_{\text{COOH}} = \frac{1}{1 + 10^{\text{pH} - \text{pKa}_1}}
  $$

- **Fraction of Deprotonated Form** ($f_{\text{COO}^-}$):

  $$
  f_{\text{COO}^-} = \frac{10^{\text{pH} - \text{pKa}_1}}{1 + 10^{\text{pH} - \text{pKa}_1}}
  $$

**Î±-Amino Group**:

- **Fraction of Protonated Form** ($f_{\text{NH}_3^+}$):

  $$
  f_{\text{NH}_3^+} = \frac{1}{1 + 10^{\text{pH} - \text{pKa}_2}}
  $$

- **Fraction of Deprotonated Form** ($f_{\text{NH}_2}$):

  $$
  f_{\text{NH}_2} = \frac{10^{\text{pH} - \text{pKa}_2}}{1 + 10^{\text{pH} - \text{pKa}_2}}
  $$

*Note*: For the amino group, although it is basic, the Henderson-Hasselbalch equation is applied in the same way.

Now, we'll calculate the charge... Note that the net charge of alanine at any pH is the sum of the charges from its ionizable groups:

- **Protonated Î±-Amino Group** ($\text{NH}_3^+$): $+1$ charge
- **Deprotonated Î±-Amino Group** ($\text{NH}_2$): $0$ charge
- **Protonated Î±-Carboxyl Group** ($\text{COOH}$): $0$ charge
- **Deprotonated Î±-Carboxyl Group** ($\text{COO}^-$): $-1$ charge

The **net charge** ($Z$) is calculated as:

$$
Z = (f_{\text{NH}_3^+} \times (+1)) + (f_{\text{NH}_2} \times 0) + (f_{\text{COOH}} \times 0) + (f_{\text{COO}^-} \times (-1))
$$

This simplifies to:

$$
Z = f_{\text{NH}_3^+} - f_{\text{COO}^-}
$$

The pH at which the net charge = 0 is know as the isoelectric point $p_i$.

Finally, we'll use the following Python program to Visualize the Relationship and determine the isoelectric point.

Notice how we break up the problem, and make two plots by simply running the plotting procedure twice. 

````python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Define the pKa values for alanine
pKa1 = 2.34   # Î±-carboxyl group
pKa2 = 9.46   # Î±-amino group

# Define the pH range
pH = np.linspace(0, 14, 1000) # creates an array to hold 1000 increments of pH scale, from 0-14. 

# Calculate the fraction of protonated and deprotonated forms for each group

# a-Carboxyl Group
ratio_carboxyl = 10 ** (pH - pKa1)
f_COOH = 1 / (1 + ratio_carboxyl)            # Fraction of protonated form (COOH)
f_COO_minus = ratio_carboxyl / (1 + ratio_carboxyl)  # Fraction of deprotonated form (COO^-)

# a-Amino Group
ratio_amino = 10 ** (pH - pKa2)
f_NH3_plus = 1 / (1 + ratio_amino)  # Fraction of protonated form (NH3+)
f_NH2 = ratio_amino / (1 + ratio_amino)     # Fraction of deprotonated form (NH2)

# Calculate the net charge at each pH
net_charge = f_NH3_plus * (+1) + f_NH2 * (0) + f_COOH * (0) + f_COO_minus * (-1)
# This is more information than we need, but I want to include all the ionization states to highlight the concept.
# This also simplifies to: net_charge = f_NH3_plus - f_COO_minus

# Plot the fraction of each species
plt.figure(figsize=(10, 6)) # 10 and 6 are the size dimensions of the figure. Feel free to change these and see what happens.

# Plot fractions for the a-carboxyl group
plt.plot(pH, f_COOH, label='Protonated Î±-Carboxyl (COOH)', linestyle='--')
plt.plot(pH, f_COO_minus, label='Deprotonated Î±-Carboxyl (COOâ»)', linestyle='--')

# Plot fractions for the a-amino group
plt.plot(pH, f_NH3_plus, label='Protonated Î±-Amino (NHââº)', linestyle='-.')
plt.plot(pH, f_NH2, label='Deprotonated Î±-Amino (NHâ)', linestyle='-.')

plt.xlabel('pH')
plt.ylabel('Fractional Abundance')
plt.title('Fraction of Protonation States of Alanine vs. pH')
plt.legend()
plt.grid(True) # change this to false and see what happens...
plt.show()

# Plot the net charge
plt.figure(figsize=(10, 6))
plt.plot(pH, net_charge, label='Net Charge of Alanine', color='purple')
plt.axhline(0, color='grey', linestyle='--')
plt.xlabel('pH')
plt.ylabel('Net Charge')
plt.title('Net Charge of Alanine as a Function of pH')
plt.legend()
plt.grid(True)
plt.show()
````
